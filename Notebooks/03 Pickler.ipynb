```python
from pickle import Unpickler
from matplotlib import pyplot as plt

ap_archivo = open ("datos_matrices.dat", "rb")
pick = Unpickler(ap_archivo)

while True:
    try:
        data = pick.load()
        plt.matshow(data)
    except EOFError:
        break
```

```python
from matplotlib import cm #Colormap module

ap_archivo2 = open ("datos_matrices2.dat", "rb")
pick = Unpickler(ap_archivo2)

data = pick.load()

fig,axis = plt.subplots(1,len(data.keys()))

for ikey,iaxis in zip(data.keys(),axis):
    iaxis.imshow(data[ikey],cmap=cm.RdBu)
```
