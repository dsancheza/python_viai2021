## Introducción a python (3) - Numpy, Scipy y Matplotlib

Las librerias numpy, scipy y matplotlib son las líbrerias básicas para el desarrollo de aplicaciones científicas en python.

**numpy** permite crear arreglos de n-dimensiones (vectores, matrices, cubos...) y define las operaciones básicas entre ellos.

### Creación de Arreglos con numpy

Las funciones arange y linspace permiten crear arreglos de valores entre dos valores.

**arange** es una extensión de la función range que regresa un arreglo en lugar de una lista.

**linspace (a,b,n)** crea un espacio lineal de n elementos en el rango [a,b]. Nótese que a diferencia de range y arange, **linspace** si incluye en valor b.

```python
import numpy as np

v1 = np.arange(-2,2,0.5)
v2 = np.linspace (-2,2,100)

print ("--------v1-------")
print (v1)
print ("--------v2-------")
print (v2)
```

Numpy define las operaciones básicas como operaciones elemento a elemento:

```python
x = np.arange(-2,2,0.01)
x2 = x**2

print (x)
print ("---X^2---")
print (x2)
print ("---log(X/X^2)--")
print (np.log(x/x2))
print ("---cos(X)----")
print (np.cos(x))
```

Podemos utilizar **matplotlib** para visualizar nuestros resultados:

```python
from matplotlib import pyplot as plt
%matplotlib inline 
plt.figure()
plt.plot (x, x**2)
plt.plot (x, np.log(x/x**2))
plt.plot (x, np.cos(x))
```

Podemos mejorar un poco la gráfica agregando una leyenda

```python
from matplotlib import pyplot as plt
%matplotlib inline 
plt.figure()
plt.plot (x, x**2, label="x^2")
plt.plot (x, np.log(x/x**2), label="log(x/x^2)")
plt.plot (x, np.cos(x), label = "cos(x)")
plt.legend()
```

Podemos indicarle a matplotlib que use latex para las leyendas:

```python
from matplotlib import pyplot as plt
%matplotlib inline 
plt.figure()
plt.plot (x, x**2, label="$x^2$")
plt.plot (x, np.log(x/x**2), label=r"$\log (x/{x^2})$")
plt.plot (x, np.cos(x), label = "$\cos(x)$")
plt.legend()
```

Agregamos las etiquetas de los ejes y un título:

```python
from matplotlib import pyplot as plt

plt.figure()
plt.plot (x, x**2, label="$x^2$")
plt.plot (x, np.log(x/x**2), label="$\log (x/{x^2})$", ls = "--", lw=3)
plt.plot (x, np.cos(x), label = "$\cos(x)$", ls = "-.")
plt.xlabel ("$x (s)$")
plt.ylabel ("$f(x) (V)$")
plt.suptitle(u"La gráfica para el artículo")
plt.legend()
```

### Operaciones con matrices

Podemos generar valores o espacios lineales en dos dimensiones con la función meshgrid

```python
x = np.linspace(-20,20,10)
y = np.linspace (0,10,10)

xx,yy = np.meshgrid (x,y)


plt.matshow(xx)
plt.colorbar()
plt.matshow(yy)
plt.colorbar()
matx = np.cos(xx)+np.sin(yy)
plt.matshow(matx)
plt.colorbar()
```

**numpy** tiene un tipo de datos especial llamado matrix, el cual incluye operaciones como la transpuesta (T), la matrix inversa (I).

```python
rmtx = np.random.normal(size=[20,20])
rmtx = np.matrix(rmtx)
plt.matshow(rmtx.T)
plt.matshow(rmtx.I)
plt.matshow(rmtx*rmtx.I)      #Esto debe de dar la identidad
```

Los arreglos de numpy pueden calcular algunos valores estadíticos de manera intrínseca con los métodos max(), min(), mean() y std(). Sin embargo otros valores como la mediana se llaman como funciones de la librería

```python
print ("Máximo de matriz:  ", matx.max())
print ("Mínimo de matriz:  ", matx.min())
print ("Promedio de matriz:", matx.mean())
print ("Desviación Std     ", matx.std())
print ("Mediana de matriz  ", np.median(matx))
```

Los valores estádisticos de orden superior necesitan de la libreria de **scipy**

```python
from scipy import stats

print ("Oblicuidad de matriz:  ", stats.skew(matx))
print ("Curtosis de matriz:    ", stats.kurtosis(matx, axis=None))
```

### Persistencia (Guardar y cargar los datos)

Tenemos varias opciones:

  **Numpy** (loadtxt, savetxt). Estas funciones guardan/cargan desde/hacia archivos de texto.
  
  
  **Pickle** (pickle.Pickler, pickle.Unpickler.  Estas funciones guardan/cargan desde/hacia archivos binarios.
  

Por ejemplo, guardemos los resultados de las gráficas:    

```python
x = np.arange(-2,2,0.01)

cuadrado =  x**2
logaritmico = np.log(x/x**2)
coseno = np.cos(x)

np.savetxt ("datos_graficas.txt", [x,cuadrado,logaritmico, coseno])
```

```python
from pickle import Pickler
ap_archivo = open ("datos_matrices.dat", "wb")
pick = Pickler(ap_archivo)
pick.dump (rmtx)
pick.dump (rmtx.T)
pick.dump (rmtx.I)
ap_archivo.close()
```
